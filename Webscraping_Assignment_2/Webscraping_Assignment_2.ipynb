##### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
#lets connect the driver
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
#opening the naukri page on automated chrome driver
driver.get("https://www.naukri.com/")
```

```python
#Entering designation in the automated web page through chrome driver 
#find element using class name
desig=driver.find_element(By.CLASS_NAME,'suggestor-input')
#putting the value in place of designation in the automated web page
desig.send_keys("Data Analyst")
```

```python
#Entering location in the automated web page through chrome driver
#find element using absolute xpath
loc=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")  
loc.send_keys("Bangalore")
```

```python
#click into the submit button
submit=driver.find_element(By.CLASS_NAME,"qsbSubmit")
submit.click()
```

**now in the automated web page we have our searched data that has to be scraped.**

```python
# Making empty list
job_title=[]
job_location=[]
company_name=[]
expeerience_required=[]
```

```python
#finding elements by relative xpath for job title
title=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')

#iterate data into for loop to append into job_title list
for i in title:
    job_title.append(i.text)

# slicing of data to get first 10 job_title
job_title=job_title[0:10]
print("no of element present inside the job_title list= ",len(job_title))
print(job_title)
```

```python
#finding elements by relative xpath for job locations
loc=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')

#iterate data using for loop to job_location list using append function.
for i in loc:
    job_location.append(i.text)
    
# slicing the list to get first 10 job location
job_location=job_location[0:10]
print("length of list, job_location =",len(job_location))
print(job_location)
```

```python
# finding elements by relative xpath for company name
name=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')

# iterate data using 'for loop' 
for i in name:
    company_name.append(i.text)
    
#slicing the data 
company_name=company_name[0:10]
print("length of list,company_name= ",len(company_name))
print(company_name)
```

```python
# finding elements using relative x path for 'experience required'
exp=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')

#iterate data into the "for loop" to get required elements
for i in exp:
    expeerience_required.append(i.text)


#slicing to get required data
expeerience_required=expeerience_required[0:10]
print("lenth of expeerience_required list = ",len(expeerience_required))
print(expeerience_required)
```

```python
#Finally making a dataframe of the scraped data.
df1=pd.DataFrame({"Job Title":job_title,"Job Location":job_location,"Company Name":company_name,"Experience Required":expeerience_required})

#calling the dataframe
df1
```

```python
#to quit driver of the automated web page.
driver.quit()
```

##### Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:
https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART
As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100reviews.

```python
#loading driver and webpage
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market")
```

```python
#creating empty list
rating=[]
summary=[]
review=[]

a=0 #initialization
while a<=10:
    #find elements of rating using relative xpath
    rat=driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    time.sleep(3)
    for i in rat:
        rating.append(i.text)
    
    
    #finding elements of review summary
    summ=driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    time.sleep(3)
    for j in summ:
        summary.append(j.text)
    
    
    #finding elements of Full Review
    rev=driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    time.sleep(3)
    for k in rev:
        review.append(k.text)
    
    
    #click the next button
    nxt=driver.find_elements(By.XPATH,'//a[@class="_1LKTO3"]')
    nxt[-1].click()
    time.sleep(10)
    
    #increment for the next loop
    a+=1

#slicing operation   
rating=rating[0:100]
summary=summary[0:100]
review=review[0:100]

print("rating=",len(rating),"summary=",len(summary),"review=",len(review))
print(rating)
print(summary)
print(review)
```

```python
driver.quit()
```

```python
#creating dataframe for review
df5=pd.DataFrame({"Rating":rating,"Review Summary":summary,"Full Review":review})
df5
```

#### Q8: Write a python program to scrape data for Top 1000 Quotes of All Time. The above task will be done in following steps:
1. First get the webpage https://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
#loading the driver and get the webpage
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.azquotes.com/")
```

```python
#click on Top Quotes
top_quotes=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
top_quotes.click()
```

```python
#Create empty list
quote=[]
author=[]
type_of_quote=[]

for x in range(1,11):
    qts=driver.find_elements(By.XPATH,'//div[@class="wrap-block"]')
    for i in qts:
        quote.append(i.text.split('\n')[0])
        author.append(i.text.split('\n')[1])
        type_of_quote.append(i.text.split('\n')[2])
    #tap on next button
    if x==10:
        break
    nxt=driver.find_elements(By.XPATH,'//li[@class="next"]')
    nxt[-1].click()
    time.sleep(5)
    
    
    
print("quote=",len(quote),"author",len(author),"Type of quote=",len(type_of_quote))
print(quote,"\n\n")
print(author,"\n\n")
print(type_of_quote)
```

```python
driver.quit()
```

```python
#creating dataframe of quotes
df8=pd.DataFrame({"QUOTE":quote,"AUTHOR":author,"TYPE OF QUOTE":type_of_quote})
df8
```
