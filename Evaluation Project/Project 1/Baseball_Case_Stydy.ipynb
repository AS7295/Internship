## Data Preprocessing

```python
# Checking for outliers
from scipy.stats import zscore
z=np.abs(zscore(df))
```

```python
# as per Emperical Rule thresold limit for outliers is 3
thresold=3
np.where(z>thresold)
```

```python
# cross Checking the z score of perticular element
z.iloc[5,1]
```

Hence it is clear that the z value of index [5,1] is greater than 3. Hence we need to remove that perticular row for the better model preperation.

```python
# Removing Outliers
df=df[(z<3).all(axis=1)]
df
```

```python
# Checking shape of the dataset after removing the dataset.
df.shape
```

Hare we can see that there is only one row is removed from the dataset.

```python
# checking Skewness
df.skew()
```

The permissible limit for skewness is from -0.5 to +0.5. Hare we can notice that there are skewness present in H, CG, SHO, SV and E column.

```python
# removing skewness by cuberoot transformation
df['H']=np.cbrt(df['H'])
df['CG']=np.cbrt(df['CG'])
df['SHO']=np.cbrt(df['SHO'])
df['SV']=np.cbrt(df['SV'])
df['E']=np.cbrt(df['E'])
```

```python
# Checking skewness wheather skewness has removed or not
df.skew()
```

Still there are skewness present in some column like H, CG and E column.

```python
figrow=int(np.ceil(len(df.columns)/3))
plotnum=1
plt.figure(figsize=(figrow*2,5*3))
for i in df.columns:
    if plotnum<=len(df.columns):
        plt.subplot(figrow,3,plotnum)
        sns.kdeplot(data=df, x=i)
        plt.xlabel(i,fontsize=10)
    plotnum+=1
plt.tight_layout()
```

## Feature Engineering

```python
# Seperating Features and Label from the dataset
x=df.drop("W", axis=1)
y=df['W']
print("Shape of Features", x.shape)
print("Shape of Labels",y.shape)
```

```python
# Feature scaling using standard scaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x.head()
```

```python
# Checking Variance Inflation Factor(VIF)
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['Freatures']=x.columns
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif
```

VIF values causing multi coliniearity problem.Hare we can find the features ER, ERA, and AB is greater than 10 which means they have high correlation with other features. We need to drop the column which have high value of vif values.

```python
# Dropping out the ER column
x.drop("ER",axis=1, inplace=True)
x.head()
```

```python
# again looking for vif values
vif=pd.DataFrame()
vif['Freatures']=x.columns
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif
```

Now there is multicolinearity present in between RA and ERA column. hence we need to drop the column which have higher vif values.

```python
# droping ERA column
x.drop("ERA", axis=1, inplace=True)
```

```python
vif=pd.DataFrame()
vif['Freatures']=x.columns
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif
```

Now we can notice that the vif values of each column is less than 10. Hence Multicolinearity issue has been resolved. Now we can look for Model Preparation.

## Model Preparation

```python
# importing libraries
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, ExtraTreesRegressor, AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
# Finding the best Random State
maxAccu=0
maxRS=0
for i in range(200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    accu=r2_score(y_test,pred)
    if accu>maxAccu:
        maxAccu=accu
        maxRS=i
print(f"Maximum r2score is {maxAccu} is at {maxRS}")
```

```python
#Spliting the dataset into train data and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=maxRS)
print(f"The shape of training data is {x_train.shape} and \nThe shape of test data is {x_test.shape}.")
```
